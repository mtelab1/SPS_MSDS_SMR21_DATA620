# Data Cleaning

```python
# Import required libraries
import networkx as nx
import networkx.algorithms.bipartite as bipartite
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pyvis.network import Network
from datetime import datetime
from scipy.stats import zscore
from tabulate import tabulate
from networkx.drawing.nx_pydot import graphviz_layout
```

```python
#Import Dataset
df = pd.read_csv('ia-wiki-user-edits-page.edges', sep = ' ',header = None, names = ['source','target','weight','time'])
```

```python
#Convert timestamp to datetime format
df['date'] = [datetime.utcfromtimestamp(v).strftime('%Y-%m-%d') for v in df['time']]
```

```python
df['hour'] = pd.to_datetime(df['date']).dt.strftime("%H")
```

```python
#tag nodes to distiguish later
df['source'] = ['u'+ str(v) for v in df['source']]
df['target'] = ['p'+ str(v) for v in df['target']]
```

```python
#Filter to include only latest years data
df['date'] = pd.to_datetime(df['date'])
maxyear = df['date'].dt.year.max()
df = df[df['date'].dt.year == maxyear]
print(len(df))
```

```python
#With repeat edits common, we can roll the values to a single edge and assign the count to the 'weight' attribute
df_grouped = df.groupby(['source','target']).aggregate({('weight'):np.sum}).reset_index()
```

## Island Method
While still in the pandas format, we will enlist the island method by filtering the edges to include only most active users and pages; while simultaniuosly removing the uber active users, which we have already determined are bots.

```python
#Set "water levels"
freq_std = freq_stats['std']
user_min = user_stats['75%']
page_min = page_stats['75%']
#apply filter
df_island = df_grouped[(df_grouped['pages']>user_min)  & (df_grouped['users']>page_min) & (df_grouped['frequency'] < (freq_std * 5))]
```

```python
print('From the above steps, we have reduced a list of almost', len(df), 'edges to', len(df_island)) 
```

```python
#modifiction of code clock found on Social Network Analysis for Startups, pg64 
def trim_edges(g, weight=1):
    l = []
    g2 = g.copy(as_view=False)
    for f, to, edata in g.edges(data=True):
            if edata['weight'] < weight:
                l.append((f,to))
    g2.remove_edges_from(l)
    g3 = nx.subgraph(g2, g2.nodes())
    return g3
```

```python
# We will create a function that will take the dataframe as input and the filter params and create the graph and calcualte the 
# centrality metrics and display as output

def get_centrality_ranks(Graph, rows):
    
    f=dict(nx.degree(Graph))
    fc_sorted_tuples = sorted(f.items(), key=lambda item: item[1], reverse=True)
    fc_sorted_dict = {k: v for v,(k,r) in enumerate(fc_sorted_tuples)}
    
    d=nx.degree_centrality(Graph)
    dc_sorted_tuples = sorted(d.items(), key=lambda item: item[1], reverse=True)
    dc_sorted_dict = {k: v for v,(k,r) in enumerate(dc_sorted_tuples)}
    
    e=nx.eigenvector_centrality(Graph)
    ec_sorted_tuples = sorted(e.items(), key=lambda item: item[1], reverse=True)
    ec_sorted_dict = {k: v for v,(k,r) in enumerate(ec_sorted_tuples)}
    
    b=nx.betweenness_centrality(Graph)
    bc_sorted_tuples = sorted(b.items(), key=lambda item: item[1], reverse=True)
    bc_sorted_dict = {k: v for v,(k,r) in enumerate(bc_sorted_tuples)}
    
    c=nx.closeness_centrality(Graph)
    cc_sorted_tuples = sorted(c.items(), key=lambda item: item[1], reverse=True)
    cc_sorted_dict = {k: v for v,(k,r) in enumerate(cc_sorted_tuples)}

    ## use Python sets to compute a union of the sets 
    #names=list(set(names1) | set(names2)| set(names3)| set(names4)  )
    ## use the networksx vorterrank function to get a sorted list of nodes based on its influence ranking:
    names =nx.voterank(Graph)
    ## build a table with centralities 
    
    table=[[name,fc_sorted_dict[name],dc_sorted_dict[name],ec_sorted_dict[name],bc_sorted_dict[name],cc_sorted_dict[name]] for name in names]
    col_names = ["Node", "Degrees","Degree_Centrality", "Eigenvector", "Betweeness", "Closeness"]
    df = pd.DataFrame(table,columns = col_names)
    #print (tabulate(table, headers= col_names))
    return df
```

```python
# We will create a function that will take the dataframe as input and the filter params and create the graph and calcualte the 
# centrality metrics and display as output

def get_centrality_values(Graph, rows):
    
    f=dict(nx.degree(Graph))
    fc_sorted_tuples = sorted(f.items(), key=lambda item: item[1], reverse=True)
    fc_sorted_dict = {k: v for k, v in enumerate(fc_sorted_tuples)}
    
    d=nx.degree_centrality(Graph)
    dc_sorted_tuples = sorted(d.items(), key=lambda item: item[1], reverse=True)
    dc_sorted_dict = {k: v for k, v in enumerate(dc_sorted_tuples)}
    
    e=nx.eigenvector_centrality(Graph)
    ec_sorted_tuples = sorted(e.items(), key=lambda item: item[1], reverse=True)
    ec_sorted_dict = {k: v for k, v in enumerate(ec_sorted_tuples)}
    
    b=nx.betweenness_centrality(Graph)
    bc_sorted_tuples = sorted(b.items(), key=lambda item: item[1], reverse=True)
    bc_sorted_dict = {k: v for k, v in enumerate(bc_sorted_tuples)}
    
    c=nx.closeness_centrality(Graph)
    cc_sorted_tuples = sorted(c.items(), key=lambda item: item[1], reverse=True)
    cc_sorted_dict = {k: v for k, v in enumerate(cc_sorted_tuples)}

    ## use Python sets to compute a union of the sets 
    #names=list(set(names1) | set(names2)| set(names3)| set(names4)  )
    ## use the networksx vorterrank function to get a sorted list of nodes based on its influence ranking:
    names =nx.voterank(Graph)

    ## build a table with centralities 
    table=[[name,f[name],d[name],e[name],b[name],c[name]] for name in names]
    col_names = ["Node", "Degrees","Degree_Centrality", "Eigenvector", "Betweeness", "Closeness"]
    df = pd.DataFrame(table,columns = col_names)
    #print (tabulate(table, headers= col_names))
    return df
```
